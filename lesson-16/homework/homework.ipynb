```python
import pandas as pd

# Create a dictionary with the data
flight_data = {
    'destinations': ['New York', 'Los Angeles', 'Los Angeles', 'Miami', 'Seattle', 
                     'Miami', 'Dallas', 'Atlanta', 'Dallas', 'San Francisco'],
    'origin': ['LAX', 'JFK', 'ORD', 'MIA', 'SEA', 
               'BOS', 'DFW', 'ATL', 'DEN', 'SFO'],
    'dest': ['JFK', 'LAX', 'ORD', 'MIA', 'SEA', 
             'BOS', 'DFW', 'ATL', 'DEN', 'SFO'],
    'carrier': ['Delta', 'United', 'American', 'Southwest', 'Alaska', 
                'JetBlue', 'Delta', 'United', 'American', 'Alaska'],
    'miles per hour': [np.nan,10,np.nan,18,23,np.nan,15,18,np.nan,11]
}

# Create the DataFrame
flights = pd.DataFrame(flight_data)

# Display the DataFrame
flights
```

```python
#1st task: generate except destinations all columns with their corresponding rows.


origin_dest_carrier=flights[['origin', 'dest', 'carrier']]
origin_dest_carrier
```

```python
flights.head(3)
```

```python
# 2nd task: choose unique destinations from flights table.

unique_destinations=flights['destinations'].unique()
number_of_unique_destinations=len(unique_destinations)
print('Number of unique destinations:', number_of_unique_destinations)
print('Unique destinations are', unique_destinations)
```
