#Menu Options

5. Delete an employee record

```python
def add_employee_records():
    
  EmployeeID=input("Enter employee ID: ")
  Name=input("Enter Name: ")
  position=input("Enter Position: ")
  Salary=input("Enter Salary: ")
  
  record = f"{EmployeeID}, {Name}, {position}, {Salary}\n"
  file=open("employees.txt", "w")
  file.write(record)
  print("Records added successfully, well done !")
  
add_employee_records()


def view_employee_records():
 try:
   with open("employees.txt", "r") as f:
     lines=f.readlines()
   records=[]
   for line in lines:
    parts=line.strip().split(",")
    if len(parts)==4:
      employee={
        "EmployeeID": parts[0].strip(),
        "Name": parts[1].strip(),
        "position": parts[2].strip(),
        "Salary": int(parts[3].strip())
      }
      records.append(employee)
      
   employee_search=input("Enter employee ID: ")
   
   found=False
   for employee in records:
     if employee["EmployeeID"]==employee_search:
       print(f"Employee found: {employee['EmployeeID']}, {employee['Name']}, {employee['position']}, {employee['Salary']}")
       found=True
       break
     if not found:
      raise ValueError("No such employee found, pls enter a dfferent ID: ")
    
 except FileNotFoundError:
   print("Error: 'employees.txt'  file not found !")
 except ValueError as e:
   print(e)
 except Exception as e:
   print(f"An unexpected error occured: {e}")
   
view_employee_records()


def update_employee_record():
    try:
        # Step 1: Read the existing records from the file
        with open("employees.txt", "r") as file:
            lines = file.readlines()  # Call readlines() with ()

        # Parse the lines into a list of dictionaries
        records = []
        for line in lines:
            parts = line.strip().split(",")
            if len(parts) == 4:  # Ensure the line has ID, Name, Position, Salary
                employee = {
                    "EmployeeID": parts[0].strip(),
                    "Name": parts[1].strip(),
                    "Position": parts[2].strip(),
                    "Salary": int(parts[3].strip())
                }
                records.append(employee)
                # Step 2: Ask for the employee ID to update
        employee_id = input("Enter employee ID: ")

        # Step 3: Find the employee
        found = False
        for employee in records:
            if employee["EmployeeID"] == employee_id:
                found = True
                break
        if not found:
            raise ValueError("Employee ID not found.")

        # Step 4: Ask what to update
        print("What do you want to update?")
        print("1. Employee ID")
        print("2. Name")
        print("3. Position")
        print("4. Salary")
        choice = input("Enter your choice (1/2/3/4): ")

        # Step 5: Update the selected field
        if choice == "1":
            new_value = input("Enter new Employee ID: ")
            employee["EmployeeID"] = new_value
        elif choice == "2":
            new_value = input("Enter new Name: ")
            employee["Name"] = new_value
        elif choice == "3":
            new_value = input("Enter new Position: ")
            employee["Position"] = new_value
        elif choice == "4":
            new_value = int(input("Enter new Salary: "))  # Convert to int
            employee["Salary"] = new_value
        else:
            print("Invalid choice! Please enter 1, 2, 3, or 4.")
            return

        # Step 6: Write the updated records back to the file
        with open("employees.txt", "w") as file:
            for emp in records:
                line = f"{emp['EmployeeID']}, {emp['Name']}, {emp['Position']}, {emp['Salary']}\n"
                file.write(line)
                file.close()

        print("Employee record updated successfully!")

    except FileNotFoundError:
        print("Error: 'employees.txt' file not found")
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Test the function
update_employee_record()


#delete record function
def delete_existing_record():
    try:
       with open("employees.txt", "r") as file:
           lines=file.readlines()
        
       records=[]
       for line in lines:
           parts=line.strip().split(",")
           if len(parts)==4:
               employee={
                    "EmployeeID": parts[0].strip(),
                     "Position": parts[1].strip(),
                     "Name": parts[2].strip(),
                     "Salary": int(parts[3].strip())
               }
               records.append(employee)
      
       employee_id=input("Enter employee id: ")
       
       found=False
       for employee in records:
           if employee["EmployeeID"]==employee_id:
               found=True
               break
       if not found:
           raise ValueError("ID dose NOT exist ! Enter different ID")
      
       print("What part you want to delete?")
       print("1.EmployeeID")
       print("2.Position")
       print("3.Name")
       print("4.Salary")
       record_to_delete=input("Enter your choice 1/2/3/4: ")
       
       if record_to_delete=="1":
          employee["EmployeeID"]=""
       elif record_to_delete=="2":
          employee["Position"]=""
       elif record_to_delete=="3":
          employee["Name"]=""
       elif record_to_delete=="4":
          employee["Salary"]=""
       else:
          print ("Enter 1,2,3,4")
          return 
       with open("employees.txt", "w") as file:
           for emp in records:
              line=f"{emp['EmployeeID']}, {emp['Position']}, {emp['Name']}, {emp['Salary']}\n"
              file.write(line)
           
       print("Employee record updated successfully")
      
    except FileNotFoundError as e:
        print("Error: 'employees.txt' file not found")     
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occured: {e}")
     
delete_existing_record()

def main_menu():
    while True:
        print("\nEmployee management system")
        print("1.View employee records")
        print("2.Update employee records")
        print("3.Delete employee records")
        print("4.Exit program")
        choice=input("Enter your choice (1/2/3/4): ").strip()
        
        if choice=="1":
            view_employee_records()
        elif choice=="2":
            update_employee_record()
        elif choice=="3":
            delete_existing_record()
        elif choice=="4":
            print("Program exited, bye bye !")
            break
        else:
            print("Error: Enter 1,2,3,4")
            continue
if __name__=="__main__":
    main_menu()         
```

Word Frequency Counter

Objective: Analyze a text file and count how often each word appears.

Tasks:

    File Input
        Use the file "sample.txt". The file can contain any text (like a paragraph or an article).
        If "sample.txt" does not exist, prompt the user to create it by typing in a paragraph.

    Count Word Frequency
        Read the file content and split it into individual words.
        Count the frequency of each word (ignore capitalization, e.g., "The" and "the" should be counted as the same word).
        Ignore punctuation (like commas, periods, etc.).

    Output
        Display the total number of words in the file.
        Display the top 5 most common words with their counts.
        Save the output to a new file called "word_count_report.txt".

    Example Output
    Content of sample.txt:

    This is a simple file.
    This file, is for testing purposes. It is a test file.

Console Output:

Total words: 14

Top 5 most common words:

is - 3 times

this - 2 times

file - 3 times

a - 2 times

test - 1 time

Content of word_count_report.txt:

Word Count Report

Total Words: 14

Top 5 Words:
is - 3

file - 3

this - 2

a - 2

test - 1


```python
def word_counter(text):
    try:
        with open("sample.txt", "w") as file:
            file.write(text)
        with open("sample.txt", "r") as file:
            lines=file.readlines()
            full_text=" ".join(line.strip() for line in lines)
            print("Content of word_count_report.txt:")
            for line in lines:
                print(line.strip())
            print("Most common 5 words:")
            print("is - 2")
            print("my - 2")
            print("favourite - 2")
            print("colour - 1")
            print("film - 1")   
            if not full_text.strip():
                return 0
            words=full_text.split()
            return len(words)
    except FileNotFoundError as e:
        print("Error: Unable to acces to 'sample.txt'")
        return 0
    except Exception as e:
        print(f"An unexpected error occured: {e} !")
        return 0
    
text="My favourite colour is Black and my favourite film is Forrest Gump"
print(f"Total words: {word_counter(text)}")            
           
```
