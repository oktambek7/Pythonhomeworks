1. Basic Plotting

    Task: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.


$f(x) = x^2 - 4x + 4$

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10, 10, 1000)
y=x**2-4*x+4

plt.plot(x, y, c='b', label=("$ f(x) = x^2 - 4x + 4 $"))
plt.title('Parabola')
plt.xlabel('x axis', c='g')
plt.ylabel('y axis', c='r')
plt.legend()
plt.show()
```

2. Sine and Cosine Plot

    Task: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.


```python
from math import sin, cos
x=np.linspace(0, 2*np.pi, 100)

sin_x=np.sin(x)
cos_x=np.cos(x)

plt.figure(figsize=(4, 4))
plt.title('Sin and Cos functions')
plt.xlabel('x', c='purple')
plt.ylabel('y', c='g', rotation=0)
plt.grid(True)
plt.plot(x, sin_x, 'r-', linewidth=2, label='sin(x)')
plt.plot(x, cos_x, 'b--o', linewidth=1, label='cos(x)', markersize=5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random x and y values from a uniform distribution between 0 and 10
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Define a list of colors and markers to cycle through
colors = ['red', 'blue', 'green', 'purple', 'orange']
markers = ['o', 's', '^', 'D', '*']

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set figure size
for i in range(100):
    plt.scatter(x[i], y[i], 
                c=colors[i % len(colors)],  # Cycle through colors
                marker=markers[i % len(markers)],  # Cycle through markers
                s=50)  # Size of points

# Add title, axis labels, and grid
plt.title('Scatter Plot of 100 Random Points', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

5. Histogram

    Task: Generate a random dataset of 1000 values sampled from a normal distribution (mean=0, std=1). Plot a histogram of the data with 30 bins. Add a title and axis labels. Adjust the transparency of the bars using the alpha parameter.


```python
data=np.random.randn(1000)
np.mean(data)==0
np.std(data)==1
plt.hist(data, bins=30, color='purple', alpha=0.7, width=1, edgecolor='k')
plt.title('Histogram', color='b')
plt.xlabel('x axis')
plt.ylabel('y axis', rotation=0)
plt.grid(True, linestyle='--')
plt.show()
```

6. 3D Plotting

    Task: Create a 3D surface plot for the function $ f(x, y) = \cos(x^2 + y^2) $ over the range of $ x $ and $ y $ values from -5 to 5. Use a suitable colormap and add a colorbar. Set appropriate labels for the axes and title.


```python
fig=plt.figure(figsize=(6, 5))
ax=plt.axes(projection='3d')

x=np.linspace(-5, 5, 1000)
y=np.linspace(-5, 5, 1000)
X, Y=np.meshgrid(x, y)

Z=np.cos(X**2 + Y**2)

surf=ax.plot_surface(X, Y, Z, cmap='viridis')

plt.colorbar(surf, ax=ax, label='f(x, y)')

ax.set_xlabel=('x')
ax.set_ylabel=('y')
ax.set_zlabel=('z')
ax.set_title('Plot of $f(x, y)=x^2 + y^2$')

plt.show()
```
